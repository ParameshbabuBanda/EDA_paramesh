# Empirical rule

- mean-1*std to mean+1*std : 68%
- mean-2*std to mean+2*std : 95%
- mean-3*std to mean+3*std : 99.7%

```python
# step-1: calculate the meann of prevaialing wage
# step-2: calculate the std of prevailig wage
# step-3: lb=mean-1*std
# step-4: ub=mean+1*std
# step-5: con1= data<ub
# step-6: con2= data>lb
# step-7: con3= con1 and con2
# step-8: data[con3]
# step-9: compare with 68% 25480 count
```

```python

mean = visa_df['prevailing_wage'].mean()
std = visa_df['prevailing_wage'].std()
lb = mean-1*std
ub = mean+1*std
con1=visa_df['prevailing_wage'] < ub
con2=visa_df['prevailing_wage'] > lb
con3= con1 & con2
visa_df[con3]
```

```python
data=visa_df['prevailing_wage']
for i in range(1,4):
    mean = visa_df['prevailing_wage'].mean()
    std = visa_df['prevailing_wage'].std()
    lb = mean-i*std
    ub = mean+i*std
    con1=visa_df['prevailing_wage'] < ub
    con2=visa_df['prevailing_wage'] > lb
    con3= con1 & con2
    if i==1:
            print(f'This is 68% of data count : {len(data[con3])}')
    elif i==2:
            print(f'This is 95% of data count : {len(data[con3])}')
    else:
            print(f'This is 99.7% of data count : {len(data[con3])}')

    
```
